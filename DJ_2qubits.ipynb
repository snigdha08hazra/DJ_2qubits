## Constant : for 2 qubits [f(x) = 0]

```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram

def Deutsch_Josza_algorithm(oracle_gate):
    # Create a circuit with three qubits and two classical bits
    qc = QuantumCircuit(3, 2)
    
    # Apply X gate to the last qubit
    qc.x(2)
    
    # Apply Hadamard gates to all three qubits
    qc.h(range(3))
    
    # Append the oracle gate
    qc.append(oracle_gate, range(3))
    
    # Apply Hadamard gates to the first two qubits
    qc.h(range(2))
    
    # Measure the first and second qubits and store the results in the first and second classical bits
    qc.measure([0, 1], [0, 1])
    
    # Draw the circuit
    qc.draw('mpl')
    
    return qc

def oracle_constant():
    qc = QuantumCircuit(3)
    return qc

oracle_gate_constant = oracle_constant().to_gate()
oracle_gate_constant.name = "Constant function"
dj_circuit_constant = Deutsch_Josza_algorithm(oracle_gate_constant)
aer_sim = Aer.get_backend('aer_simulator')
transpiled_dj_circuit_constant = transpile(dj_circuit_constant, aer_sim)
qobj_constant = assemble(transpiled_dj_circuit_constant)
results_constant = aer_sim.run(qobj_constant).result()
answer_constant = results_constant.get_counts()
plot_histogram(answer_constant)
```
